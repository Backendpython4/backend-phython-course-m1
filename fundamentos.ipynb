## Función id()

### id() función para obtener la dirección de memoria

```python
# Cadena / string
my_string = "python"
print(my_string)
print(id(my_string))
```

```python
# Cadena / string
my_string2 = "Backend"
print(my_string2)
print(id(my_string2))
```

```python
# Cadena / string
my_string = "python"
my_string3 = "python"
print(my_string)
print(my_string3)
print(type(my_string))
print(type(my_string3))
print(id(my_string))
print(id(my_string3))
```

## Mútiples Valores de Entrada

```python
a,b,c = input("Ingresas los 3 valores: ").split()
print("El valor de a es: ",a)
print("El valor de a es: ",b)
print("El valor de a es: ",c)
```
